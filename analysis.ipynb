```python
import wandb

# Initialize the W&B API
api = wandb.Api()
wandb.login()

# Specify the project and entity (username or team name)
entity = "hbin0701"  # Replace with your W&B username or team name
project = "Arith-Transfer-Multi"        # Replace with your project name

# Load the project
runs = api.runs(f"{entity}/{project}")
```

```python
from tqdm.auto import tqdm

all_dict = {}

for run in tqdm(runs):
    name = run.name
    history = run.history(samples=50000) 
    if name.lower() == 'task1':
        break
    for x, y in zip(history.index, history['test_accuracy']):
        epoch = x + 1
        if y > 0.95:
            all_dict[name] = epoch
            break
    
    if name not in all_dict:
        all_dict[name] = 999999
```

```python
def organize_tasks_table(all_dct):
    # Initialize a dictionary to store organized tasks
    organized = {}
    
    # First pass: Handle single tasks
    for key, value in all_dct.items():
        if '->' not in key:
            organized[key] = {
                'count': value,
                'incoming_paths': []
            }
    
    # Second pass: Handle task paths
    for key, value in all_dct.items():
        if '->' in key:
            source, target = key.split('->')
            if target not in organized:
                organized[target] = {
                    'count': 0,
                    'incoming_paths': []
                }
            organized[target]['incoming_paths'].append({
                'from': source,
                'count': value
            })
    
    # Print table header
    print(f"{'Task':<10} {'Original Training':<20} {'Continual Training':<35} {'% Change'}")
    print("-" * 80)
    
    # Print organized results as table
    for task in sorted(organized.keys()):
        base_count = organized[task]['count']
        incoming = organized[task]['incoming_paths']
        
        # Print the main task row with original count
        print(f"{task:<10} {base_count:<20}", end='')
        
        # Handle tasks with no incoming paths
        if not incoming:
            print("".ljust(35), "N/A")
            continue
            
        # Print first incoming path
        first_path = incoming[0]
        percent_change = ((first_path['count'] - base_count) / base_count * 100) if base_count > 0 else float('inf')
        percent_str = f"{percent_change:+.1f}%" if base_count > 0 else "N/A"
        print(f"{first_path['from']} → {task} ({first_path['count']})".ljust(35), end='')
        print(f"{percent_str}")
        
        # Print remaining incoming paths
        for path in incoming[1:]:
            percent_change = ((path['count'] - base_count) / base_count * 100) if base_count > 0 else float('inf')
            percent_str = f"{percent_change:+.1f}%" if base_count > 0 else "N/A"
            print(f"{'':<30}{path['from']} → {task} ({path['count']})".ljust(35), end='')
            print(f"{percent_str}")

organize_tasks_table(all_dict) -
```
